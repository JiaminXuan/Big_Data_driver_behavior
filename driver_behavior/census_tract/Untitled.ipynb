```python
from pandas import DataFrame as df
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
count=df.from_csv('/Users/xuanjm/GitHub/big-data-project/output/trip_dur-hist.csv',header=None,sep='\t')
count=count.sort_index()
count.columns=['counts']
```

```python
count.index.name='hours'
```

```python
%matplotlib inline
count.iloc[:140].plot(kind='bar',figsize=(15,9),title='Histogram of trip duration')
```

```python
count['hours']=count.index
```

```python
count.hours=map(int,count.hours)
```

```python
count.index=range(len(count.index))
```

```python
hours=count.groupby(by='hours').sum()
```

```python
hours.loc[:15].plot(kind='bar')
```

```python
hours['hours']=hours.index
```

```python
hours.to_csv('hourbyour',index=False)
```
